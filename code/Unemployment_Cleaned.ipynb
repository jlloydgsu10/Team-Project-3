```python
import pandas as pd
```

```python
unemployment = "static/data/UNRATE.csv"
```

```python
unemployment_df = pd.read_csv(unemployment)
unemployment_df
```

```python
unemployment_df['DATE'] = pd.to_datetime(unemployment_df['DATE'])
unemployment_df = unemployment_df.loc[unemployment_df["DATE"]>"2009-01-01"]
unemployment_df
```

```python
# unemployment_df.to_csv("UNRATE_cleaned.csv", index=True)
```

```python
oil_file = "static/data/oil_price.csv"
oil_price_df = pd.read_csv(oil_file)
oil_price_df['Date'] = pd.to_datetime(oil_price_df['Date'],errors="coerce")

oil_price_df.sort_values(by='Date')
# oil_price_df1= oil_price_df.rename(columns={"date":"Date", "value": "oil_price"})
# count = oil_price_df['date'].value_counts()
# count
oil_price_df
```

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots()
ax1.set_title("Yearly Unemployment Rate vs Oil Price", fontweight='bold', fontsize = '20')
ax1.bar(unemployment_df['DATE'], unemployment_df['UNRATE'],color='purple', alpha = 0.5, width = 20, label = 'Unemployment Rate')

# ax1.bar(us_oilres_df['year'], us_oilres_df['us_reserves'],color='red', alpha = 0.4, width = 200, label = 'US Oil Reserves')
# ax1.set_xlabel('Date', fontweight='bold', fontsize = '20')
# # ax1.set_ylim(-25, 50)
ax1.set_ylabel('Unemployment Rate Average by Year', fontweight='bold', fontsize = '20')
ax1.grid(True, color = 'k', linestyle= ":")
ax1.legend(loc=2)

ax2 = ax1.twinx()
ax2.scatter(oil_price_df['Date'], oil_price_df['oil_price'],marker='o',color='red', label = 'Oil Price')
ax2.set_ylabel('Oil Price', fontweight='bold', fontsize = '20')
# # plt.xticks(rotation = 45)
# ax2.set_ylim(-50, 150)

plt.figure(figsize=(15, 8))
rcParams['figure.figsize']= 18,10
# ax2.legend(loc=1)
```
